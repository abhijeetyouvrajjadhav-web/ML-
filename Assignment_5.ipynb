```python
import matplotlib.pyplot as mp
import numpy as np

years = ['2001', '2006', '2010', '2015', '2018']
sachin_runs = [1600, 1300, 1562, 0, 0]
kohli_runs = [0, 887, 995, 1380, 1325]
sehwag_runs = [850, 900, 1100, 0, 0]

x = np.arange(len(years))
width = 0.25

mp.figure(figsize = (10,6))
mp.bar(x - width, sachin_runs, width, label = 'Sachin')
mp.bar(x,kohli_runs, width, label = 'Kohli')
mp.bar(x + width, sehwag_runs, width, label = 'Sehwag')

mp.xlabel('YEARS')
mp.ylabel('RUNS')
mp.title('Over 5 selected Years')
mp.xticks(x, 'YEARS')

mp.legend()
mp.show() 
```

Q.no.5

Bar Chart with Annotations: Draw a vertical bar chart for five categories of your choice. Annotate the value on top of each bar.

```python
import matplotlib.pyplot as abhi 
import numpy as np

players=['Dhoni','Sachin','ABD','Virat','Rohit']
runs=[16266,34357,11014,29000,18000]
colours=['#D1CE0D','#55A1FF','#256400','#9D0000','#004DAB']

abhi.figure(figsize=(8,7))
bars=abhi.bar(players,runs,color=colours)

for bar in bars:
    yval = bar.get_height()
    abhi.text(bar.get_x() + bar.get_width()/2, yval + 0.5, str(yval), ha='center', va='bottom')

abhi.xlabel("PLAYERS NAME")
abhi.ylabel("Total intl runs")
abhi.title("Runs of 5 legend players" )

abhi.grid(axis='y',linestyle='--',alpha=0.7)
abhi.show() 
```

Q.no.6.

Multiple Lines in One Plot: Plot three mathematical functions (sin, cos, exp) on the same axes, each with different colors and line styles. Add a legend.

```python
import matplotlib.pyplot as abhi
import numpy as np

x = np.linspace(0, 2 * np.pi, 100)

y_sin = np.sin(x)
y_cos = np.cos(x)
y_exp = np.exp(x/3) #avoid steep growth

abhi.figure(figsize=(8, 6))
abhi.plot(x, y_sin, color='blue', linestyle='-', label='sin(x)')
abhi.plot(x, y_cos, color='green', linestyle='-.', label='cos(x)')
abhi.plot(x, y_exp, color='red', linestyle=':', label='exp(x/3)')

abhi.xlabel("x")
abhi.ylabel("y")
abhi.title("Multiple Functions: sin(x), cos(x), exp(x/3)")
abhi.legend()
abhi.grid(True)

abhi.show()
```

Q.no.8

Subplots Grid: Create a 2x2 grid of subplots: line plot, bar chart, scatter plot, and histogram, each in a different subplot. Add an overall (supertitle) for the figure.

```python
import matplotlib.pyplot as abhi
import numpy as np

x_line = np.linspace(0, 10, 100)
y_line = np.sin(x_line)

Animes = ['One Piece', 'Demon slayer', 'Bleach', 'Pokemon', 'DBZ']
networth = [21, 15, 7.5, 92, 28]

x_scatter = np.random.rand(50)
y_scatter = np.random.rand(50)

hist_data = [96, 82, 75, 91, 67, 88, 59, 3, 84, 62, 
        90, 77, 68, 94, 56, 7, 7, 91, 70, 85, 
        92, 66, 74, 0, 97, 94, 60, 76, 83, 
        93, 78, 71, 58, 86, 72, 3, 69, 65, 98, 
        57, 99, 55, 81, 53, 52, 54, 1, 50, 100]

fig, axs = abhi.subplots(2, 2, figsize=(12, 10))

# Line plot
axs[0, 0].plot(x_line, y_line, color='#398DF5')
axs[0, 0].set_title("Line Plot")

# Bar chart
axs[0, 1].bar(Animes, networth, color='orange')
axs[0, 1].set_title("Animes Networth in Billion $ :(Bar Chart)")

# Scatter plot
axs[1, 0].scatter(x_scatter, y_scatter, color='green',s=50)
axs[1, 0].set_title("Scatter Plot")

# Histogram
axs[1, 1].hist(hist_data, bins=5, color='purple', edgecolor='black')
axs[1, 1].set_title("Histogram")

# Supertitle
fig.suptitle("2x2 Grid of Various Plots", fontsize=16)

# Adjust layout to prevent overlap
fig.tight_layout(rect=[0, 0.03, 1, 0.95])

# Show the plot
abhi.show() 
```

Q.no.9

Stack Plot: Plot the cumulative sales of 3 products over 12 months as a stack plot using made-up data.

```python
import matplotlib.pyplot as abhi
import numpy as np

months = np.arange(1, 13)

product_A = [10, 15, 10, 18, 22, 25, 28, 30, 35, 38, 40, 45]
product_B = [5, 8, 12, 10, 14, 16, 18, 10, 22, 24, 26, 28]
product_C = [2, 4, 6, 7, 8, 9, 10, 12, 13, 14, 16, 18]

abhi.figure(figsize=(9, 5))
abhi.stackplot(months, product_A, product_B, product_C,
              labels=['Oppo', 'Samsung', 'others'],
              colors=['#e4ecf1', '#1F7DF0', '#F0EDAB'])

abhi.xlabel("Month")
abhi.ylabel("Sales")
abhi.title("Cumulative Sales of 3 Products Over 12 Months")
abhi.legend(loc='upper left')
abhi.grid(True, linestyle='--', alpha=0.7)

abhi.show()
```

Q.no.11

Axis and Legend Customization: For a random line plot, reverse the x-axis, move the legend outside the plot, and increase legend font size.

```python
import matplotlib.pyplot as abhi
import numpy as np

x = np.arange(0, 10, 1) #0-9
y1 = np.random.randint(10, 50, size=10)
y2 = np.random.randint(10, 60, size=10)

# Plot lines
abhi.plot(x, y1, label='Luffy', color='#045BC6')
abhi.plot(x, y2, label='Zoro', color='#BB3305')

# Reverse x-axis
abhi.gca().invert_xaxis() # gca=gets current axis

abhi.title("Random Line Plot with Custom Axis and Legend")
abhi.xlabel("X-axis")
abhi.ylabel("Y-axis")

# Move legend outside the plot (right side) & increase font size
abhi.legend(loc='center left', bbox_to_anchor=(1, 0.9), fontsize='large')

abhi.tight_layout()

abhi.show()
```

Q.no.12

Pie Chart with Explode: Create a pie chart showing the market share of four companies. Explode the largest segment and display percentage values on the chart.

```python
import matplotlib.pyplot as abhi

companies = ['Apple', 'Samsung', 'Xiaomi', 'Others']
market_share = [35, 30, 10, 15]
colours=['#e4ecf1','#1F7DF0','#F8B16A','#96CAEB']

explode = [0.1, 0, 0, 0] # 'explode' Apple

abhi.figure(figsize=(8, 6))
abhi.pie(market_share, labels=companies,
        explode=explode,
        colors=colours,
        shadow=True,
        autopct='%1.1f%%',
        startangle=90)

abhi.title('Smartphone Market Share')

abhi.axis('equal')

abhi.show()
```
